# Simple notebook to do a comparitive time series between one or more countries

```python
import requests
import io 
import pandas as pd
url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))

print (df["Country/Region"])
print (df.head)

```

# Fetch data for India 

```python
df_india = df.loc[df['Country/Region'] == "India"]
df_india = df_india.transpose()
df_india=df_india.drop(df_india.index[[0,1,2,3]])
# iterating the columns 
df_india.columns=['India']
df_india.head()
#df_india.plot(figsize=(25,10))

```

```python
df_hk = df.loc[df['Province/State'] == "Hong Kong"]

df_hk = df_hk.transpose()
df_hk=df_hk.drop(df_hk.index[[0,1,2,3]])
df_hk.columns=['HK']
df_hk.head()
```

```python
df_hubei = df.loc[df['Province/State'] == "Hubei"]

df_hubei = df_hubei.transpose()
df_hubei=df_hubei.drop(df_hubei.index[[0,1,2,3]])
df_hubei.columns=['Hubei']
df_hubei.head()
```

```python
df_sgp = df.loc[df['Country/Region'] == "Singapore"]

df_sgp = df_sgp.transpose()
df_sgp=df_sgp.drop(df_sgp.index[[0,1,2,3]])
df_sgp.columns=['Singapore']
df_sgp.head()
```

```python
df_iran = df.loc[df['Country/Region'] == "Iran"]
df_iran = df_iran.transpose()
df_iran=df_iran.drop(df_iran.index[[0,1,2,3]])
df_iran.columns=['Iran']
df_iran.head()
```

```python
df_france = df.loc[df['Country/Region'] == "France"]
df_france = df_france.transpose()
df_france=df_france.drop(df_france.index[[0,1,2,3]])
df_france['France'] =  df_france.iloc[: , 0]

# Iterate over the sequence of column names to append
for index in range(df_france.shape[1]-1):
   if index>0:
       df_france['France'] =  df_france['France'] +  df_france.iloc[: , index]

# Remove the extra columns 
for index in range(df_france.shape[1]-1):
    df_france.drop(df_france.columns[0], axis = 1, inplace = True) 
    
df_france.head()
```

```python
df_usa = df.loc[df['Country/Region'] == "US"]
df_usa = df_usa.transpose()
df_usa=df_usa.drop(df_usa.index[[0,1,2,3]])
df_usa['USA'] =  df_usa.iloc[: , 0]

# Iterate over the sequence of column names to append
for index in range(df_usa.shape[1]-1):
   if index>0:
       df_usa['USA'] =  df_usa['USA'] +  df_usa.iloc[: , index]

# Remove the extra columns 
for index in range(df_usa.shape[1]-1):
    df_usa.drop(df_usa.columns[0], axis = 1, inplace = True) 
    
df_usa.head()

```

```python
df_china = df.loc[df['Country/Region'] == "China"]
df_china = df_china.transpose()
df_china=df_china.drop(df_china.index[[0,1,2,3]])
df_china['China'] =  df_china.iloc[: , 0]

# Iterate over the sequence of column names to append
for index in range(df_china.shape[1]-1):
   if index>0:
       df_china['China'] =  df_china['China'] +  df_china.iloc[: , index]

# Remove the extra columns 
for index in range(df_china.shape[1]-1):
    df_china.drop(df_china.columns[0], axis = 1, inplace = True) 
    
df_china.head()

```

```python


df_korea = df.loc[df['Country/Region'] == "Korea, South"]
df_korea = df_korea.transpose()
df_korea=df_korea.drop(df_korea.index[[0,1,2,3]])
df_korea.columns=['Korea']
df_korea.head()
```

```python
df_germany = df.loc[df['Country/Region'] == "Germany"]
df_germany = df_germany.transpose()
df_germany=df_germany.drop(df_germany.index[[0,1,2,3]])
df_germany.columns=['Germany']
df_germany.head()
```

```python
df_italy = df.loc[df['Country/Region'] == "Italy"]

df_italy = df_italy.transpose()
df_italy=df_italy.drop(df_italy.index[[0,1,2,3]])
df_italy.columns=['Italy']


df_italy.head()
```
